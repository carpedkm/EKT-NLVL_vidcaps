```python
# for saat ft extraction
gt_anno_pth = '/saat/charades_train_mapping_for_vid_cap.json'
with open(gt_anno_pth, 'r') as f:
    gtanno = json.load(f)
```

```python
# import unsup bd composite event
```

```python
with open('/saat/unsup event/composite_events_unsup_ft_charades.pickle', 'rb') as f:
    tmp = pickle.load(f)
```

```python
# filter only the training ones
with open('/saat/charades_train_mapping_for_vid_cap.json', 'r') as f:
    for_fil = json.load(f)
```

```python
training_keys = list(for_fil.keys())
```

```python
# get gtanno as {vidID : {duration: <>, segmentcnt: <>}}
gtanno_reph = {}
for key in list(gtanno.keys()):
    bd = gtanno[key]
    segmentcount = len(bd['segment'])
    duration = bd['duration']
    gtanno_reph[key] = {'duration' : duration, 'segmentcount' : segmentcount}
```

# match the number with the charades original dataset

```python
from tqdm import tqdm
cnt = 0
new_dic_for_unsup = {} # {keys:{'duration':value, 'segment':segment}}
cnt = 0
for iter in tqdm(range(100)):
    print(cnt)
    print('iterations', iter)
    keylist = list(tmp.keys())
    random.shuffle(keylist)
    for key in keylist:
        # print(tmp[key])
        bd_info = tmp[key]
        random.shuffle(bd_info)
        if key[:-4] in list(gtanno_reph.keys()):
            get_gtanno = gtanno_reph[key[:-4]]
        else:
            continue
        # print(bd_info)
        duration = get_gtanno['duration']
        if iter < len(bd_info):
            if cnt == 9190:
                break
            to_append_bdinfo_start = bd_info[iter][0] * duration
            to_append_bdinfo_end = bd_info[iter][1] * duration
            tmpseg = [to_append_bdinfo_start, to_append_bdinfo_end]
            

            if key in list(new_dic_for_unsup.keys()):
                new_dic_for_unsup[key[:-4]]['segment'].append(tmpseg)
            else:
                new_dic_for_unsup[key[:-4]] = {'segment' : [tmpseg], 'duration' : duration}
            cnt += 1
            # print(bd_info)
            # print(tmp[key])
            # print(bd_info[iter])
            bd_info.remove(bd_info[iter])
            tmp[key] = bd_info
            # print(bd_info)
            # print(tmp[key])
            # break
        else:
            pass
    # break
    if cnt == 12418:
        break
        
```

```python
new_dic_for_unsup
```

```python
with open(unsup_pth, 'w') as f:
    json.dump(new_dic_for_unsup, f)
```
