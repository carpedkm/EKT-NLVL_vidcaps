# make c3d (feat 3d) h5py

```python
from tqdm import tqdm
import os
```

```python
_3dfeat_path = '/saat/ActivityNet_for_SAAT/Feature_3D'
temp_load = os.listdir(_3dfeat_path)
```

```python

no_key = []
f = h5py.File('/saat/gtboundary_anet_h5py/_anet_c3d.h5', 'w')
for key in temp_load:
    try:
        feat = np.load(os.path.join(_3dfeat_path, key))
        f.create_dataset(str(key[:-4]), data=feat.mean(0), dtype='f8')
    except:
        no_key.append(key)
f.close()
```
